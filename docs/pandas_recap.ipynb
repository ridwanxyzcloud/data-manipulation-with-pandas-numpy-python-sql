### Joins or merge

```python
customer_df
```

```python
sales_df
```

```python
product_df = pd.DataFrame({'product':['A','B','C'], 'category':['Finance','Investment','Insurance']})
product_df
```

```python
# establishing a relationship between the two df using 'join'
#inner join

joined_df = pd.merge(sales_df, product_df, on='product', how='inner')
joined_df
```

```python
#left join

left_join_df = pd.merge(sales_df, product_df, on='product', how='left')
left_join_df
```

```python
#right join

right_join_df = pd.merge(sales_df,product_df, on='product', how='right')
right_join_df
```

```python
#right join

right_join_df = pd.merge(sales_df, customer_df, on='customer_id', how='right')
right_join_df
```

```python
# outer join
#right join

outer_join_df = pd.merge(sales_df, product_df, on='product', how='outer')
outer_join_df
```
